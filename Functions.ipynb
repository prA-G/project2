## 6. Explain the concept of generator in python and how are they defined?

A generator in Python is a special type of iterator that allows you to iterate through a sequence of values, but doesnnot store them all in memory at once.

Generators are defined just like normal functions, but they use the yield statement instead of return

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```

```python
gen=count_up_to(3)
print(next(gen))
```

```python
print(next(gen))
```

```python
print(next(gen))
```

## 7. What are advantages of using generators over regular functions?

1. Great for handling large datasets or streams of data.
2. It starts giving results immediately without waiting for the entire process to complete.
3. You can create infinite data streams without crashing your system.
4. Generator syntax (yield) keeps your code concise and easier to read.
5. Generators internally handle iteration logic.

```python
def simple_gen():
    yield "First"
    yield "Second"

gen = simple_gen()
print(next(gen))
print(next(gen))
```

## 9. Explain the purpose and uses of map function in Python.

- Def : The map() function in Python is used to apply a specific function to each item in an iterable (like a list, tuple, etc.) and returns a map object (an iterator).
- Purpose :
1. To transform or process all elements of an iterable without using a loop.
2. It makes code shorter, faster, and more readable.
- Uses :
1. Works with functions (including lambda functions)
2. Supports lazy evaluation (returns an iterator)



```python
def square(x):
    return x * x

nums = [1, 2, 3, 4]
result = map(square, nums)
print(list(result))
```

## 10. What is difference between map(), reduce(), and filter()?

- map() : Applies a function to each element of an iterable and returns a new iterable (map object) with the results.
- reduce() : Applies a function cumulatively to reduce the iterable to a single value.
- filter() : Filters elements of an iterable based on a condition (function that returns True/False).

```python
#map()
nums = [1, 2, 3]
squared = list(map(lambda x: x * x, nums))
print(squared)
```

```python
#reduce()
from functools import reduce
nums = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, nums)
print(total)
```

```python
#filter()
nums = [1, 2, 3, 4]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

## 1. Write a Python function that takes a list of number as input and return the sum of all even number in list.

```python
def sum(numbers):
    sum1 = 0
    for num in numbers:
        if num % 2 == 0:
            sum1 += num
    return sum1
```

```python
my_list = [10, 21, 4, 45, 66, 93]
result = sum(my_list)
print("Sum of even numbers:", result)
```

## 4. Write a python function that check if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
```

## 6. Write a generator function in Python that yeild the power of two up to a given exponent.

```python
def power_of_two_gen(limit):
    for exponent in range(limit + 1):
        yield 2 ** exponent
```

```python
max_exponent = 5

print(f"\nPowers of 2 from 2^0 to 2^{max_exponent}:")
for value in power_of_two_generator(max_exponent):
    print(value)
```

## 7. Implement a generator function that read a file line by line and yeild each line as a string.

```python
def read(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

## 10. Create a python program that use filter() to remove all the vowels from given string.

```python
def remove_vow(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, text))

input_word="DataScience"

result = remove_vow(input_word)
print("String without vowels:", result)
```
